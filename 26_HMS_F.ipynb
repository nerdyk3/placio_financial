```python
import networkx as nx #Graphical persentation of unshorted data
import numpy as np #calculation lib
import pandas as pd #data Shoting Lib eg:series
import matplotlib.pyplot as mlt #for plot the data 
mlt.style.use('fivethirtyeight')
import seaborn as sns #statical data vertualisation 
import plotly.offline as py #plotly is a data_virtualisation 
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls

#check_out and ls will help in the to show files and folders

from subprocess import check_output
print(check_output(["ls"]).decode("utf8"))
```

```python
HMS27_ = pd.read_csv('csv/26_july.csv')
HMS27_
```

```python
D26_ = HMS27_.sort_values(by='Collected Amount' , ascending= False)
D26_
```

```python
mlt.subplots(figsize=(10,7))
sns.countplot(x=D26_['Mode'],hue=D26_['Mode'], data=D26_).set_title("Tenant count Payment Mode wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(40,9))
sns.countplot(x=D26_['Property Name'],hue=D26_['Property Name'], data=D26_).set_title("Tenant count property wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(20,7))
sns.countplot(x=D26_['Mode'],hue=D26_['Term'], data=D26_).set_title("monthly and yearly count payment mode wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(46,9))
sns.countplot(x=D26_['Collected Amount'],hue=D26_['Property Name'], data=D26_).set_title("amount collected property wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(50,9))
sns.countplot(x=D26_['Property Name'],hue=D26_['Status'], data=D26_).set_title("status Property wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(80,7))
sns.countplot(x=D26_['Tenant Code'],hue=D26_['Mode'], data=D26_).set_title("Payment Mode used buy tenant")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(80,7))
sns.countplot(x=D26_['Tenant Code'],hue=D26_['Term'], data=D26_).set_title("Tenant term used")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```
