```python
month_july = pd.read_csv('csv/tbl_tenant_collect_fee.csv')
```

```python
sortDateElement=month_july.sort_values(by='create_date',ascending=False)
julyDate=sortDateElement[1:400]
julyDate
```

```python
julyDate['status'].unique()
```

```python
mlt.subplots(figsize=(20,18))
sns.countplot(x=julyDate['status'],hue=julyDate['status'], data=julyDate).set_title("status 2018-07-01 to 2018-07-23")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
#mlt.subplots(figsize=(20,18))
#sns.countplot(x=julyDate['due_amount'],hue=julyDate['tenant_id'], data=julyDate).set_title("status 2018-07-01 to 2018-07-23")
#mlt.savefig('status_20180701_to_20180723.png')
#mlt.show()
x = julyDate['tenant_id']
y = julyDate['collect_fee']
z = julyDate['due_amount']

fig=mlt.figure(figsize=(400.,10.))
fig.subplots_adjust(left=0.1,right=0.9,bottom=0.1,top=0.9)

ax=fig.add_subplot(111)
#ax = mlt.subplot(111)
ax.bar(x, y,width=5,color='b',align='center')
ax.bar(x, z,width=5,color='g',align='center')
ax.set_xticklabels(julyDate['tenant_id'])
mlt.ylabel("due and collected ammount")
mlt.xlabel("tenant_id")
mlt.title("collect_and_due_amount_20180701_to_20180723 & Green for Due Amount and Blue for Collect Fee")
#mlt.savefig('collect_and_due_amount_20180701_to_20180723.png')
mlt.show()
```
