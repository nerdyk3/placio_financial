```python
import networkx as nx #Graphical persentation of unshorted data
import numpy as np #calculation lib
import pandas as pd #data Shoting Lib eg:series
import matplotlib.pyplot as mlt #for plot the data 
mlt.style.use('fivethirtyeight')
import seaborn as sns #statical data vertualisation 
import plotly.offline as py #plotly is a data_virtualisation 
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls

#check_out and ls will help in the to show files and folders

from subprocess import check_output
print(check_output(["ls"]).decode("utf8"))
```

```python
modeR = pd.read_csv('tbl_tenant_collect_fee.csv')
```

```python
modeR
```

```python
modeR['payment_mode']
```

```python
mlt.subplots(figsize=(20,18))
sns.countplot(x=modeR['payment_mode'],hue=modeR['payment_mode'], data=modeR).set_title("Graph of modes using by tenants until 21/07/2018")
#mlt.savefig('modes_tenants_until_21072018.png')
mlt.show()
```

```python
b = modeR[modeR.status > 0]
b
```

```python
mlt.subplots(figsize=(20,18))
sns.countplot(x=b['status'],hue=b['status'], data=b).set_title("Graph for Paid/Pending until 21/07/2018")
#mlt.savefig('Paid_Pending.png')
mlt.show()
```

```python
d = b.loc[:,['tenant_id','collect_fee','due_amount']]
d
```

```python
mlt.subplots(figsize=(200,18))
sns.lineplot(data=d.collect_fee, color="g" )
ax2 = mlt.twinx()
sns.lineplot(data=d.due_amount, color="b", ax=ax2).set_title("blue for DUE AMOUNT and Green for COLLECTED AMOUNT")
mlt.savefig('due_and_collect_amount Graph.png')
mlt.show()
```

```python
x=d.describe()
x.to_html('table.html')
```

```python
d['due_amount'].groupby(d['collect_fee'].d.month).sum().plot.bar()
```

```python
mlt.subplots(figsize=(20,18))
sns.barplot(x=d['tenant_id'], y=d['collect_fee'], hue=d['collect_fee'], data=d).set_title("Graph for Paid/Pending until 21/07/2018")
#mlt.savefig('Paid_Pending.png')
mlt.show()
```
